```python
# It is mainly used in
 
# Clustering similar documents
# Clustering customers based on similar features
```

```python
from sklearn.datasets import make_blobs
# The make_blobs functions is used to generate synthetic datasets for clusterring and classification tasks
# This function will create clusters of data points with "Gaussian Distribution"
```

```python
#creating random dataset
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=5.0,random_state=101)
```

```python
# n_samples=Total number of points equally divided among clusters.
# n_featuers=It indicated the number of features(columns)
# centers=It determine number of clusters to be generated
# cluster_std=It sets the standers deviation of the clusters, High value makes the clusters to sperad out
data
```

```python
import matplotlib.pyplot as plt
x,y=data
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')       #Here c=y is a inbulit funcyion to get the colour to culters
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot for k means")
```

```python
data[0].shape
```

```python
from sklearn.cluster import KMeans
k=KMeans(n_clusters=4)
```

```python
k.fit(data[0])
```

```python
k.cluster_centers_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=k.labels_,s=50,edgecolor='black')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],s=50)
```

```python
# Diff between KNN and K means  Clustering 

#1.) 

#KNN is used for Classification and regression 
#K means is used for Clustering problems 

#2.)

#KNN is supervised lgorithm
#K means is unsupervised algorithm

#3.)

#To training KNN, we need a dataset with all the data points having class labels
# For training K means,we no need any such information 

#4.)

#we use KNN to predict the class label or new points 
#we use K means to find patterns in a given dataset by grouping datapointsintoclusters
```
